```python
import pandas as pd
```

Excelファイルの読み込み。  
header=[0, 1]で、1行目と2行目をヘッダーに指定。  
index_col=[0, 1]で、1列目と2列目をインデックス列に指定   

```python
df = pd.read_excel('data/sample3.xlsx', header=[0, 1], index_col=[0, 1])
```

```python
df
```

Pandasでは、NaNがある列はfloatになってしまう。intにはできない。そもそもExcelの数値はすべてfloatで持っているので気にしない。

計の行と列が邪魔になるので除く  
最終行や最終列は-1で指定できる  
例: 次の式は最終セルの値を表示

```python
df.iat[-1, -1]
```

今回は、最後の2行及び最後の2列を除く必要があるので-2を指定する  
以下の式で、最後の2行、最後の2列を除いた配列を取得できる  
cross_df.iloc[0:-2, 0:-2] の省略型です。

```python
df = df.iloc[:-2, :-2]
```

```python
df
```

イテレータを使って1列に並べる  
Qiitaの記事にあわせたので、iterrowsを使っている。  
カラムで廻す場合は、iteritemsを使う

```python
sr = pd.Series()
for index_id, row in df.iterrows():
    row.index = row.index.map(lambda x, i=index_id: i + x)
    sr = sr.append(row)
```

```python
sr
```

NaNのデータは不要なので削除

```python
sr = sr.dropna()
```

```python
sr
```

インデックスがtuppleになっているので、マルチインデックスに戻す  

```python
sr.index = pd.MultiIndex.from_tuples(sr.index)
```

```python
sr
```

Excelに書き出し  
前の行と同じ項目名の場合は空白になります。  
人間が見る場合は、その方が見やすいのですが。。。  

```python
sr.to_excel('data/sample3_result.xlsx')
```

クリップボードにペーストできます
クリップボードの場合は、項目名に空白ができずにQiitaの記事と同じ形式になります

```python
sr.to_clipboard()
```

Qiitaの記事と同じ形式でExcelに出力したい場合

```python
df2 = pd.DataFrame(sr)
```

```python
df2 = df2.reset_index()
```

```python
df2.to_excel('data/sample3_result2.xlsx', header=False, index=False)
```
